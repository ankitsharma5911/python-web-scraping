```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome()  # Make sure ChromeDriver is installed
wait = WebDriverWait(driver, 10)
url="https://www.naukri.com/job-listings-sales-logistics-professional-cita-united-kingdom-uk-0-to-3-years-050825925939"
driver.get(url)
desc_elem = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"styles_JDC__dang-inner-html__h0K4t")))
description = desc_elem.text
print(description)

# time.sleep(3)
driver.quit()
```
